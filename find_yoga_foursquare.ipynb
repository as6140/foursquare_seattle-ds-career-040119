```python
import json
import requests

# Load secrets from credentials.json
url = 'https://api.foursquare.com/v2/venues/explore'
with open('.secrets/credentials.json') as f:
    params = json.load(f) 
```

```python
params['v'] = '20180323' # 'Version', code written as of documentation on march 23, 2018, API changes won't break code
params['ll'] = '47.608,-122.336', #lat/long
params['query'] = 'food', 
params['limit'] = 100 #how many I want back

#LIST WHAT YOU WANT IN YOUR SEARCH
```

```python
response = requests.get(url=url, params=params) #encoding params into the URL
data = json.loads(response.text)

#NO FILES or NOTEBOOKS SHOULD EVER CONTAIN KEYS, which the below calls will display

#params
#response.url
```

```python
results = data['response']['groups'][0]['items']
```

```python
venues = []
venue_names = []
for result in results:
    venues.append(result['venue'])
    venue_names.append(result['venue']['name'])
```

```python
import pymongo
mc = pymongo.MongoClient()
db = mc['foursquare']
coll = db['pilates']

coll.insert_many(venues)
#MONGODB IS LONG TERM STORAGE vs. DATAFRAMES ARE SHORT TERM
#STORE A DF OF INTEREST INTO MONGODB, RELOAD IT FAST LATER ON IN DIFFERENT PLACES
```
